```python
import tensorflow as tf
import torch
import subprocess
import os

def check_gpu_setup():
    """Check GPU availability and setup for both PyTorch and TensorFlow"""
    print("=== GPU Setup Diagnostic Tool ===\n")
    
    # Check NVIDIA Driver
    try:
        nvidia_smi = subprocess.check_output("nvidia-smi", shell=True)
        print("✅ NVIDIA Driver is installed and working!\n")
        print("GPU Information:")
        print(nvidia_smi.decode())
    except:
        print("❌ NVIDIA Driver not found or not working")
        print("Please install NVIDIA drivers from: https://www.nvidia.com/download/index.aspx\n")
    
    # Check CUDA availability
    print("\n=== CUDA Status ===")
    if torch.cuda.is_available():
        print(f"✅ CUDA is available")
        print(f"CUDA Version: {torch.version.cuda}")
        print(f"GPU Device: {torch.cuda.get_device_name(0)}")
        print(f"Number of GPUs: {torch.cuda.device_count()}")
    else:
        print("❌ CUDA is not available for PyTorch")
    
    # Check TensorFlow GPU
    print("\n=== TensorFlow GPU Status ===")
    print("Available GPUs:", tf.config.list_physical_devices('GPU'))
    
    # Test GPU with PyTorch
    print("\n=== PyTorch GPU Test ===")
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"PyTorch is using: {device}")
    if torch.cuda.is_available():
        # Create random tensors to test GPU
        x = torch.randn(1000, 1000).to(device)
        y = torch.randn(1000, 1000).to(device)
        z = torch.matmul(x, y)  # Perform a GPU operation
        print("✅ PyTorch GPU test successful!")
    
    # Test GPU with TensorFlow
    print("\n=== TensorFlow GPU Test ===")
    if tf.test.gpu_device_name():
        print("✅ TensorFlow is using GPU:", tf.test.gpu_device_name())
    else:
        print("❌ TensorFlow is NOT using GPU")

if __name__ == "__main__":
    check_gpu_setup()
```

```python
import tensorflow as tf
import numpy as np

# 1. GPU Setup and Configuration
def setup_gpu():
    """Configure GPU for optimal training"""
    print("GPU Setup Status:")
    
    # Check if GPU is available
    gpus = tf.config.list_physical_devices('GPU')
    if gpus:
        try:
            # Enable memory growth
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
            print("✓ GPU found and memory growth enabled")
            print(f"✓ Number of GPUs available: {len(gpus)}")
            
            # Set memory limit (optional, adjust based on your needs)
            tf.config.experimental.set_virtual_device_configuration(
                gpus[0],
                [tf.config.experimental.VirtualDeviceConfiguration(memory_limit=5120)]  # 5GB
            )
        except RuntimeError as e:
            print(f"Error setting up GPU: {e}")
    else:
        print("No GPU found. Please check your installation.")

# 2. Create a Simple CNN Model
def create_model():
    """Create a basic CNN model for demonstration"""
    model = tf.keras.Sequential([
        # CNN layers
        tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(28, 28, 1)),
        tf.keras.layers.MaxPooling2D(),
        tf.keras.layers.Conv2D(64, 3, activation='relu'),
        tf.keras.layers.MaxPooling2D(),
        tf.keras.layers.Conv2D(64, 3, activation='relu'),
        
        # Dense layers
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# 3. Load and Prepare Data
def prepare_data():
    """Load and prepare MNIST dataset"""
    print("\nPreparing data...")
    
    # Load MNIST dataset
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
    
    # Normalize and reshape data
    x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
    x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0
    
    # Create TensorFlow datasets
    train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train))\
        .shuffle(10000)\
        .batch(32)\
        .prefetch(tf.data.AUTOTUNE)
    
    test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test))\
        .batch(32)\
        .prefetch(tf.data.AUTOTUNE)
    
    return train_ds, test_ds

# 4. Training Function
def train_model():
    """Main training function"""
    # Setup GPU
    setup_gpu()
    
    # Prepare data
    train_ds, test_ds = prepare_data()
    
    # Create model
    model = create_model()
    print("\nModel Summary:")
    model.summary()
    
    # Training callbacks
    callbacks = [
        tf.keras.callbacks.ModelCheckpoint(
            'best_model.h5',
            save_best_only=True,
            monitor='val_accuracy'
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        ),
        tf.keras.callbacks.TensorBoard(log_dir='./logs')
    ]
    
    # Train the model
    print("\nStarting training...")
    with tf.device('/GPU:0'):  # Explicitly specify GPU usage
        history = model.fit(
            train_ds,
            epochs=10,
            validation_data=test_ds,
            callbacks=callbacks
        )
    
    return model, history

# 5. Monitor GPU Usage
def print_gpu_utilization():
    """Print GPU memory usage"""
    gpus = tf.config.experimental.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                memory_info = tf.config.experimental.get_memory_info('GPU:0')
                print(f"\nGPU Memory Usage:")
                print(f"Current: {memory_info['current'] / 1024**2:.2f} MB")
                print(f"Peak: {memory_info['peak'] / 1024**2:.2f} MB")
        except:
            print("Could not get GPU memory info")

# Main execution
if __name__ == "__main__":
    print("TensorFlow version:", tf.__version__)
    print_gpu_utilization()
    model, history = train_model()
    print_gpu_utilization()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, Input

# Define input shapes for RGB and grayscale images
rgb_input_shape = (128, 128, 3)  # 3-channel RGB input
gray_input_shape = (128, 128, 1)  # 1-channel grayscale input

# Define RGB input processing path
rgb_input = Input(shape=rgb_input_shape, name='rgb_input')
rgb_model = layers.Conv2D(32, (3, 3), activation='relu')(rgb_input)
rgb_model = layers.MaxPooling2D((2, 2))(rgb_model)
rgb_model = layers.Conv2D(64, (3, 3), activation='relu')(rgb_model)
rgb_model = layers.MaxPooling2D((2, 2))(rgb_model)
rgb_model = layers.Flatten()(rgb_model)

# Define grayscale input processing path
gray_input = Input(shape=gray_input_shape, name='gray_input')
gray_model = layers.Conv2D(32, (3, 3), activation='relu')(gray_input)
gray_model = layers.MaxPooling2D((2, 2))(gray_model)
gray_model = layers.Conv2D(64, (3, 3), activation='relu')(gray_model)
gray_model = layers.MaxPooling2D((2, 2))(gray_model)
gray_model = layers.Flatten()(gray_model)

# Combine (concatenate) the two processed feature vectors
combined = layers.concatenate([rgb_model, gray_model])

# Add dense layers for classification
combined = layers.Dense(64, activation='relu')(combined)
output = layers.Dense(3, activation='softmax')(combined)  # Adjust number of classes as needed

# Define the model with two inputs and one output
model = models.Model(inputs=[rgb_input, gray_input], outputs=output)

# Compile the model
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)

# Summary of the model structure
model.summary()
```
